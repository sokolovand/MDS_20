Изучите 4 распределения из библиотеки scipy: нормальное, экспоненциальное, Стьюдента, Бета. Используя функцию плотности распредления, постройте несколько графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

```python
import numpy as np

from scipy.stats import norm
from scipy.stats import expon
from scipy.stats import t
from scipy.stats import beta

import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 100)
```

###  Нормлаьное распределение

```python
n2 = norm.pdf(x, loc=0, scale=1)
n3 = norm.pdf(x, loc=0, scale=2)
n4 = norm.pdf(x, loc=0, scale=5)
n5 = norm.pdf(x, loc=0, scale=0.5)

plt.figure(figsize=(10, 10))
plt.plot(x, n2, label='scale: 1')
plt.plot(x, n3, label='scale: 2')
plt.plot(x, n4, label='scale: 5')
plt.plot(x, n5, label='scale: 0.5')
plt.legend(prop={'size': 20})
```

```python
n6 = norm.pdf(x, loc=1, scale=1)
n7 = norm.pdf(x, loc=3, scale=1)
n8 = norm.pdf(x, loc=-1, scale=1)
n9 = norm.pdf(x, loc=-3, scale=1)
n10 = norm.pdf(x, loc=0, scale=1)

plt.figure(figsize=(10, 10))
plt.plot(x, n6, label='loc: 1')
plt.plot(x, n7, label='loc: 3')
plt.plot(x, n8, label='loc: -1')
plt.plot(x, n9, label='loc: -3')
plt.plot(x, n10, label='loc: 0')
plt.legend(prop={'size': 20})
```

* У scipy.norm есть 2 основных параметра: loc - смещение и scale - масштабирование                  
* При увеличении loc происходит перемещение максимума функции распределения вдоль оси Х на число единиц, переданных в данный параметр                 
* При увеличении scale происходит растяжение/сжатие функции распределения при неизменном положении максимума распределения. В пределе распределение стремится к дельтафункции

```python
x = np.linspace(-6, 6, 100)

n1 = expon.pdf(x, loc=0, scale=1)
n2 = expon.pdf(x, loc=3, scale=1)
n3 = expon.pdf(x, loc=5, scale=1)
n4 = expon.pdf(x, loc=-3, scale=1)
n5 = expon.pdf(x, loc=-5, scale=1)

plt.figure(figsize=(10, 10))
plt.plot(x, n1, label='loc: 0')
plt.plot(x, n2, label='loc: 3')
plt.plot(x, n3, label='loc: 5')
plt.plot(x, n4, label='loc: -3')
plt.plot(x, n5, label='loc: -5')
plt.legend(prop={'size': 20})
```

* У scipy.expon также есть 2 основных параметра: loc - смещение и scale - масштабирование

* При увеличении scale происходит растяжение/сжатие функции распределения при неизменном положении максимума распределения. В пределе распределение стремится к дельтафункции        

### Распределение Стьюдента

```python
x = np.linspace(-6, 6, 100)

n1 = t.pdf(x, 1, loc=0, scale=1)
n2 = t.pdf(x, 3, loc=0, scale=1)
n3 = t.pdf(x, 10, loc=0, scale=1)
n4 = t.pdf(x, 50, loc=0, scale=1)
n5 = t.pdf(x, 100, loc=0, scale=1)

plt.figure(figsize=(10, 10))
plt.plot(x, n1, label='Степеней свободы: 1')
plt.plot(x, n2, label='Степеней свободы: 3')
plt.plot(x, n3, label='Степеней свободы: 10')
plt.plot(x, n4, label='Степеней свободы: 50')
plt.plot(x, n5, label='Степеней свободы: 100')
plt.legend(prop={'size': 15})
```

```python
x = np.linspace(-6, 6, 100)

n1 = t.pdf(x, 1, loc=3, scale=1)
n2 = t.pdf(x, 3, loc=-3, scale=1)
n3 = t.pdf(x, 10, loc=0, scale=1)


plt.figure(figsize=(10, 10))
plt.plot(x, n1, label='loc: 3')
plt.plot(x, n2, label='loc: -3')
plt.plot(x, n3, label='loc: 0')
plt.legend(prop={'size': 15})
```

```python
x = np.linspace(-6, 6, 100)

n1 = t.pdf(x, 1, loc=0, scale=1)
n2 = t.pdf(x, 3, loc=0, scale=5)
n3 = t.pdf(x, 10, loc=0, scale=0.5)


plt.figure(figsize=(10, 10))
plt.plot(x, n1, label='scale: 1')
plt.plot(x, n2, label='scale: 5')
plt.plot(x, n3, label='scale: 0.5')
plt.legend(prop={'size': 15})
```

* У scipy.t уже есть 3 основных параметра: loc - смещение, scale - масштабирование и df - степени свободы         
* При изменении количества степеней свободы происходит приближение распределения к нормлаьному виду
* При увеличении loc происходит перемещение максимума функции распределения вдоль оси Х на число единиц, переданных в данный параметр                 
* При увеличении scale происходит растяжение/сжатие функции распределения при неизменном положении максимума распределения. В пределе распределение стремится к дельтафункции   

### Бета-распределение

```python
x = np.linspace(-1, 1, 100)

n1 = beta.pdf(x, a=90, b=10, loc=0, scale=1)
n2 = beta.pdf(x, a=50, b=50, loc=0, scale=1)
n3 = beta.pdf(x, a=10, b=90, loc=0, scale=1)


plt.figure(figsize=(10, 10))
plt.plot(x, n1, label='a: 90; b: 10')
plt.plot(x, n2, label='a: 50; b: 50')
plt.plot(x, n3, label='a: 10; b: 90')
plt.legend(prop={'size': 15})
```

```python
x = np.linspace(-2, 2, 100)

n1 = beta.pdf(x, a=90, b=10, loc=1, scale=1)
n2 = beta.pdf(x, a=50, b=50, loc=-1, scale=1)
n3 = beta.pdf(x, a=10, b=90, loc=0, scale=1)


plt.figure(figsize=(10, 10))
plt.plot(x, n1, label='a: 90; b: 10')
plt.plot(x, n2, label='a: 50; b: 50')
plt.plot(x, n3, label='a: 10; b: 90')
plt.legend(prop={'size': 15})
```

```python
x = np.linspace(-2, 2, 100)

n1 = beta.pdf(x, a=90, b=10, loc=0, scale=1)
n2 = beta.pdf(x, a=50, b=50, loc=0, scale=2)
n3 = beta.pdf(x, a=10, b=90, loc=0, scale=5)


plt.figure(figsize=(10, 10))
plt.plot(x, n1, label='a: 90; b: 10')
plt.plot(x, n2, label='a: 50; b: 50')
plt.plot(x, n3, label='a: 10; b: 90')
plt.legend(prop={'size': 15})
```

* У scipy.beta есть 4 основных параметра: loc - смещение, scale - масштабирование, a - успешные исходы, b - неудачные исходы        
* Если a > b, то максимум распределения смещается вправо, в сторону бОльшей вероятности успеха. Если a < b, то максимум распределеняи смещен ближе к 0
* При увеличении loc происходит перемещение максимума функции распределения вдоль оси Х на число единиц, переданных в данный параметр                 
* При увеличении scale происходит растяжение/сжатие функции распределения при неизменном положении максимума распределения. В пределе распределение стремится к дельтафункции
