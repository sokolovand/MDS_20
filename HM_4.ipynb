Вычислить значения первой и второй производной функции в точке х=1

```python
import sympy as sym
import numpy as np

x = sym.Symbol('x')
y = x**5 + 4*sym.sin(2*x) + sym.cos(3*x + 3)

deriv_1 = sym.diff(y, x)
deriv_1_ = deriv_1.subs(x, 1)

deriv_2 = sym.diff(deriv_1, x)
deriv_2_ = deriv_2.subs(x, 1)

print('Значение первой производной в точке х=1: ', deriv_1_.evalf(5))
print('Значение второй производной в точке х=1: ', deriv_2_.evalf(5))
```

Постройте график функции, первой и второй производной на отрезке [-5;5]

```python
import matplotlib.pyplot as plt

x = sym.Symbol('x')

y = (sym.sin(2*x +1)) ** 5
arg = np.linspace(-5, 5, 100)

deriv_1 = sym.diff(y, x)
deriv_2 = sym.diff(deriv_1, x)

value_1 = []
value_2 = []

for i in arg:
    value_1.append(deriv_1.subs(x, i).evalf())
    value_2.append(deriv_2.subs(x, i).evalf())

plt.figure(figsize=(8, 6))
plt.plot(arg, value_1, label='Первая производная')
plt.plot(arg, value_2, label='Вторая производная')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
```

Напишите код для определения производной функции ReLU.

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0.0, x)

x = np.arange(-3, 3, 0.1)
y = relu(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('ReLU Function')
plt.grid()
plt.show()
```

```python
# Из графика видно, что функция равна 0, если на вход подать <= 0. А производная от 0 равна 0. 
# Если на вход подать Х больше 0, то на выход просто передается Х
# Следовательно, зависимость линейная и выражается как y=x, а первая производная от даного выражения равна 1
# Чтобы при каждом вызове данной функции не вычислять через sympy.diff значение производной, просто ограничимся двумя условиями

def relu_derivative(x):
    # Вычислить производную функции ReLU
    # Ваш код напишите ниже и результат передайте в переменную derivative
    if x <= 0:
        derivative = 0
    else:
        derivative = 1
    return derivative

derivative = []

for i in x:
    derivative.append(relu_derivative(i))
    
plt.plot(x, derivative)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

Напишите код для определения производной сигмоиды и постарайтесь понять почему данная функция активации практически не требует дополнительных вычислительных ресурсов при расчете значений производной.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x_ = np.arange(-5, 5, 0.1)
y_ = sigmoid(x_)

# Вычисление самой производной вынесено за тело функции, чтобы уменьшить количество вычислений
# Сама функция sigmoid_derivative() возвращает значение производной в точке

x = sym.Symbol('x')
y = 1 / (1 + sym.exp(-x))
y = y.diff(x)

def sigmoid_derivative(i):
    return y.subs(x, i).evalf()

derivative = []

for i in x_:
    derivative.append(sigmoid_derivative(i))
    
plt.figure(figsize=(8, 6))  
plt.plot(x_, y_, label='Функция сигмоида')
plt.plot(x_, derivative, label='Производная сигмоиды')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Sigmoid Function')
plt.grid()
```
