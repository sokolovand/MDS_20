Иногда встречаются задачи, где необходимо не просто найти минимальное значение некоторой функции, но сделать это при некоторых условиях на переменные x. Это можно сделать знакомым нам minimize из scipy.optimize при некоторых дополнительных аргументах.

```python
from scipy.optimize import minimize
```

```python
### Задайте функцию для оптимизации из условия выше (x - вектор переменных)

### YOUR CODE HERE
def f(x):
    x0 = x[0]
    x1 = x[1]
    x2 = x[2]
    x3 = x[3]
    return x0*x1*(x2 + x3)

### Условие 1. Задайте условие следуюшим образом: функция условия должна возвращать выражение для x-C, 
### где С - ограничение снизу для данного выражения из условия задачи

### YOUR CODE HERE
def constraint(x):
    # return f(x) - 25
    return x[0]*x[1]*x[2]*x[3] - 25

### Начальное значение из условия задачи

### YOUR CODE HERE
x = [2, 2, 3, 3]
f(x)
```

```python

### Ограничения на каждую переменную x
bound = (1, 5)
boundaries = (bound, bound, bound, bound)

### Дополнительное условие
constraint1 = {'type': 'ineq', 'fun': constraint}
constraints = [constraint1]

### YOUR CODE HERE
sol = minimize(f, x, method='SLSQP', bounds=boundaries, constraints=constraints, options={'disp': True})

print(sol.fun, sol.x)
```

```python
# Проверка условия 1
a = 1
for i in sol.x:
    a *= i
    
a
```
