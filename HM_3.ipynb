Дан объект в 2D пространстве. При помощи линейных отображений:

* Уменьшить объект в два раза
* Повернуть на 130 градусов
* Отразить объект относительно прямой y=x                

Затем вычислить определитель композиции преобразований (итоговой матрицы реализующей все три отображения)

```python
import numpy as np
import matplotlib.pyplot as plt

A = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
# Увеличение в 2 раза

m1 = np.array([[0.5, 0, 0], [0, 0.5, 0], [0, 0, 0.5]])
A_x2 = A @ m1
x = A_x2[:,0]
y = A_x2[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
# Повернуть на 130 градусов против часовой вокруг оси Z

m2 = np.array([[np.cos(3 * np.pi/4), np.sin(3 * np.pi/4), 0], [-np.sin(3 * np.pi/4), np.cos(3 * np.pi/4), 0], [0, 0, 1]])
A_x2 = A @ m2
x = A_x2[:,0]
y = A_x2[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
# Поворот на 130 градусов по часовой вокруг оси Z

m3 = np.array([[np.cos(3 * np.pi/4), -np.sin(3 * np.pi/4), 0], [np.sin(3 * np.pi/4), np.cos(3 * np.pi/4), 0], [0, 0, 1]])
A_x2 = A @ m3
x = A_x2[:,0]
y = A_x2[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
# Отразить относительно прямой y=x
x_, y_ = [-200, 200], [-200, 200]


# m4 = np.array([[np.cos(np.pi/4), np.sin(np.pi/4), 0], [-np.sin(np.pi/4), np.cos(np.pi/4), 0], [0, 0, 1]])
m4 = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 0]])

A_x2 = A @ m4
x = A_x2[:,0]
y = A_x2[:,1]

plt.plot(x_, y_)
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
# Композиция преобразований
x_, y_ = [-200, 200], [-200, 200]

B = m1 @ m2 @ m4
A_3 = A @ B

x = A_3[:,0]
y = A_3[:,1]

plt.plot(x_, y_)
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

Решите систему уравнений, используя матричный подход. Необходимо составить матрицу коэффициентов и вектор правых частей.

```python
coeff = np.array([[1, 0, 1, 0],
                  [-1, 1, -2, 1],
                  [4, 0, 1, -2],
                  [-4, 4, 0, 1]])

y = np.array([2, -2, 0, 5])

x = np.linalg.inv(coeff) @ y.T
print(f'x = ', x[0])
print(f'y = ', x[1])
print(f'z = ', x[2])
print(f'w = ', x[3])
```
